## Extraindo dados dos repositórios da Amazon

```python
headers = {'X-GitHub-Api-Version': '2022-11-28'}
api_base_url = 'https://api.github.com'
owner = 'amzn' # Username que vamos pegar os dados
url = f'{api_base_url}/users/{owner}/repos'
```

```python
response = requests.get(url,headers=headers)
```

```python
response.json()
```

```python
len(response.json())
```

```python
# Autenticação

acess_token = str(TOKEN)
headers = {'Autorization': 'Bearer ' + acess_token, 'X-GitHub-Api-Version': '2022-11-28'}
```

```python
# Criando uma lista de repositórios

repos_list = []

for page_num in range(1,6):
    try:
        url_page = f'{url}?page={page_num}'
        response = requests.get(url_page, headers=headers)
        repos_list.append(response.json())

    except:
        repos_list.append(None)
```

```python
len(repos_list)
```

```python
# Extraindo dados da primeira página

repos_list[0][2]['name']
```

```python
# Criando uma lista com nomes

repos_name = []

for page in repos_list:
    for repo in page:
        repos_name.append(repo['name'])
```

```python
repos_name[:10]
```

```python
len(repos_name)
```

```python
# Selecionar os dados das linguagens de todos os repositórios

repos_language = []

for page in repos_list:
    for repo in page:
        repos_language.append(repo['language'])
```

```python
repos_name[:10]
```

```python
# Criando um Data Frame

dados_amz = pd.DataFrame()
dados_amz['repository_name'] = repos_name # Criando colunas
dados_amz['language'] = repos_language

dados_amz
```

```python
# Salvar em csv

dados_amz.to_csv('amazon.csv')
```
