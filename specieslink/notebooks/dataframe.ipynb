```python
import json
import pandas as pd
```

```python
with open('../data_processed/dados_tratados.json', 'r') as file:
    data = json.load(file)
```

```python
data_coordinates = []
data_properties = []
```

```python
for record in data:
    data_coordinates.append(record['geometry'])
    data_properties.append(record['properties'])
```

```python
df_cordenadas = pd.DataFrame(data_coordinates)
df_properties = pd.DataFrame(data_properties)
df_properties.head()
```

```python
df_combined = pd.concat([df_cordenadas, df_properties], axis=1)
df_combined.head(2)
```

```python
df_combined.columns
```

```python
colunas_vazias = df_combined.isnull().sum()
colunas_vazias_ordenadas = colunas_vazias.sort_values(ascending=False)
```

```python
media = colunas_vazias_ordenadas.median()

colunas_remover = colunas_vazias_ordenadas[colunas_vazias_ordenadas > media].index
colunas_remover
```

```python
df_cleaned = df_combined.drop(columns=colunas_remover)
```

```python
df_cleaned.columns
```

```python
df_cleaned.to_csv('../data_processed/dados_tratados.csv', index=False)
```
